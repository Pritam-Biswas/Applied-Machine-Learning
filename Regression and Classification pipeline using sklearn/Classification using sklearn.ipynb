## 1.4

```python
cat_cols=[0,2,3,5,6,8,9,11,13,14,16,18,19]

num_features=[]
cat_features=[]
for i in range(0,len(col_names)):
    if i in cat_cols:
        cat_features.append(col_names[i])
    else:
        num_features.append(col_names[i])
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

numeric_transformer_us = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python

preprocessor = ColumnTransformer(transformers=[
        ('num', numeric_transformer, num_features),
        ('cat', categorical_transformer, cat_features)])

preprocessor_us = ColumnTransformer(transformers=[
        ('num', numeric_transformer_us, num_features),
        ('cat', categorical_transformer, cat_features)])

```

```python
classifiers=["LogisticRegression", "LinearSVC", "KNeighborsClassifier"]
```

```python
classifier_models=[LogisticRegression(), LinearSVC(), KNeighborsClassifier()]
```

```python
score_list=[]

print('Model scores with scaling')
for i in range(0,3):
    sc=cross_val_score(make_pipeline(preprocessor, classifier_models[i]), X_trainv, y_trainv,cv=5)
    print(classifiers[i]+' cross validation score = '+str(np.mean(sc)))

print("\n\n")
print('Model scores without scaling')
for i in range(0,3):
    sc=cross_val_score(make_pipeline(preprocessor_us, classifier_models[i]), X_trainv, y_trainv,cv=5)
    print(classifiers[i]+' cross validation score = '+str(np.mean(sc)))

#     score_list.append(sc)
# print(np.mean(np.array(score_list)))
```

For Logistic regression, cross validation score is almost the same with or without scaling. For Linear SVC, the score significantly improves with scaling. And also for KNeighbors the cross validation score significantly increases with scaling of continuous features.

## 1.5

```python
## grid search

pipe=Pipeline([('processor', preprocessor), ('classifier', LogisticRegression())])
param_grid = [
    {'classifier' : [LogisticRegression()],
    'classifier__C':np.logspace(-3,3,20)
    },
    {'classifier' : [LinearSVC()],
     'classifier__C':np.logspace(-3,3,20)},
    {'classifier' : [KNeighborsClassifier()],
     'classifier__n_neighbors':range(5,15,2)
    }
]

clf = GridSearchCV(pipe, param_grid = param_grid, cv = 5, verbose=True)

clf.fit(X_trainv, y_trainv)

# print(clf.best_params_)
```

```python
print(clf.best_params_)
```

```python
res_df=pd.DataFrame(clf.cv_results_)
```

```python
res_df['svm']='None'
res_df['logreg']='None'
```

```python
for i in range(0,res_df.shape[0]):
    if isinstance(res_df['param_classifier'][i], LogisticRegression):
        res_df['logreg'][i]="yes"
    if isinstance(res_df['param_classifier'][i], LinearSVC):
        res_df['svm'][i]="yes"
    
```

```python
plt.plot(res_df['param_classifier__n_neighbors'],res_df['mean_test_score'])
plt.xlabel('no of neighbours')
plt.ylabel('mean test score')
plt.title('mean test score vs no of neighbours')
plt.show()
# plt.plot(res_df['param_classifier__n_neighbors'], res_df['mean_test_score'])
```

```python
plt.plot(res_df.loc[res_df['svm'] == "yes"]['param_classifier__C'], res_df.loc[res_df['svm'] == "yes"]['mean_test_score'])
plt.xlabel('C value for SVM')
plt.ylabel('mean test score')
plt.title('mean test score vs C value for SVM')
plt.show()
```

```python
plt.plot(res_df.loc[res_df['logreg'] == "yes"]['param_classifier__C'], res_df.loc[res_df['logreg'] == "yes"]['mean_test_score'])
plt.xlabel('C value for logistic regression')
plt.ylabel('mean test score')
plt.title('mean test score vs C value for logistic regression')
```

```python
print('Evaluation score of the best model on test set')
print(clf.score(X_test, y_test))
```

As we see, the test score changes from 0.785 to 0.79 with the change in random seed with K Fold shuffle, The test score also changed on changing from stratified KFold to KFold shuffle.
